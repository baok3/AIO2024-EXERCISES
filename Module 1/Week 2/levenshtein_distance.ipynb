```python
'''Viết chương trình tính khoảng cách chỉnh sửa tối thiểu Levenshtein. Khoảng cách Levenshtein thể
hiện khoảng cách khác biệt giữa 2 chuỗi ký tự. Khoảng cách Levenshtein giữa chuỗi S và chuỗi T
là số bước ít nhất biến chuỗi S thành chuỗi T thông qua 3 phép biến đổi là:

- Thêm một ký tự.
- Xóa một ký tự.
- Thay thế một ký tự bằng ký tự khác.

Khoảng cách này được sử dụng trong việc tính toán sự giống và khác nhau giữa 2 chuỗi, như
chương trình kiểm tra lỗi chính tả của winword spellchecker. Ví dụ: Khoảng cách Levenshtein
giữa 2 chuỗi "kitten" và "sitting" là 3, vì phải dùng ít nhất 3 lần biến đổi. Trong đó:

• kitten -> sitten (thay "k" bằng "s")
• sitten -> sittin (thay "e" bằng "i")
• sittin -> sitting (thêm ký tự "g")'''

def delete(char):
    return 1

def add(char):
    return 1

def substract(char_1, char_2):
    if char_1 == char_2:
        return 0
    return 1

def levenshtein_distance(target, source):
    '''Xây dựng ma trận'''
    matrix = [[0 for i in range(len(source) + 1)] for j in range(len(target) + 1)]

    # Khởi tạo ma trận
    for i in range(len(target) + 1):
        matrix[i][0] = i
    for j in range(len(source) + 1):
        matrix[0][j] = j

    # Tính toán ma trận
    for i in range(1, len(target) + 1):
        for j in range(1, len(source) + 1):
            matrix[i][j] = min(matrix[i-1][j] + delete(target[i-1]),
                               matrix[i][j-1] + add(source[j-1]),
                               matrix[i-1][j-1] + substract(target[i-1], source[j-1]))

    distance = matrix[len(target)][len(source)]

    ''' Trình bày kết quả
    + kitten -> sitten (thay "k" bằng "s")
    + sitten -> sittin (thay "e" bằng "i")
    + sittin -> sitting (thêm ký tự "g") '''

    i = len(target)
    j = len(source)

    steps = []

    while i > 0 or j > 0:
        if i > 0 and matrix[i][j] == matrix[i-1][j] + delete(target[i-1]):
            steps.append(f'Xóa ký tự "{target[i-1]}" từ vị trí {i-1}')
            i -= 1
        elif j > 0 and matrix[i][j] == matrix[i][j-1] + add(source[j-1]):
            steps.append(f'Thêm ký tự "{source[j-1]}" vào vị trí {j-1}')
            j -= 1
        else:
            if target[i-1] != source[j-1]:
                steps.append(f'Thay ký tự "{target[i-1]}" bằng ký tự "{source[j-1]}" tại vị trí {i-1}')
            i -= 1
            j -= 1

    # Vì khi trình bày, bước cuối luôn in ra trước nên phải reverse để theo tuần tự.
    steps.reverse()

    result = f'Khoảng cách chỉnh sửa tối thiểu Levenshtein giữa 2 chuỗi là: {distance} bước.\n'
    result += 'Các bước biến đổi:\n'
    for step in steps:
        result += '+ ' + step + '\n'

    return result

print(levenshtein_distance(target = 'elmets', source = 'elements'))
```
