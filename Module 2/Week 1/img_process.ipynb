```python
#! gdown 1i9dqan21DjQoG5Q_VEvm0LrVwAlXD0vB

# 12

import matplotlib.image as mpimg
import numpy as np
img = mpimg.imread('dog.jpeg')

gray_img_01 = (np.max(img, axis=2) + np.min(img, axis=2)) / 2
gray_img_01[0, 0] 

# Giải thích:

# gray_img_01 được tính bằng cách lấy trung bình của giá trị lớn nhất và nhỏ nhất trong các kênh màu RGB tại mỗi điểm ảnh.
# Giá trị tại gray_img_01[0, 0] là 102.5.
```

```python
# 13
gray_img_02 = np.mean(img, axis=2)
gray_img_02[0, 0]

# Giải thích:

# được tính bằng cách lấy trung bình của các kênh màu RGB tại mỗi điểm ảnh.
# Giá trị tại gray_img_02[0, 0] là 107.7.
```

```python
# 14
gray_img_03 = 0.21 * img[:, :, 0] + 0.72 * img[:, :, 1] + 0.07 * img[:, :, 2]
gray_img_03[0, 0]

# Giải thích:

# gray_img_03 được tính bằng cách nhân các hệ số 0.21, 0.72 và 0.07 tương ứng với các kênh màu R, G, và B tại mỗi điểm ảnh.
# Giá trị tại gray_img_03[0, 0] là 126.2.
```
